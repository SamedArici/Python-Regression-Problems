## Data Preprocessing

### Create Independent and Dependent Variables

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

### Encode Categorical Variables

#### Sex Column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,1] = le.fit_transform(x[:,1]) 

# Female = 0, Male = 1
```

#### Region column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5])],remainder='passthrough')
x = np.array(ct.fit_transform(x))

# southwest = 0 0 0 1 
# southeast = 0 0 1 0
# northwest = 0 1 0 0
# northeast = 1 0 0 0 
```

### Split the variables into the train set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

### Scale the variables

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Create Machine Learning Model

### Calculate the Model's Scores

```python
def calculate_scores(regressor,algorithm_name):
    from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
    
    global df_result,index,x_test_poly
    
    if(algorithm_name != 'poly_reg'):
        y_pred = regressor.predict(x_test)
    else:
        y_pred = regressor.predict(x_test_poly)
    
    r2 = round(r2_score(y_test,y_pred),3)
    mae = round(mean_absolute_error(y_test,y_pred),3)
    mse = round(mean_squared_error(y_test,y_pred))
    
    scores = {
        'Algorithm':algorithm_name,
        'R2 Score': r2,
        'Mean Absolute Error': mae,
        'Mean Squarred Error':mse
    }
    
    new_df = pd.DataFrame(data = scores,index = [index])
    df_result = pd.concat([df_result,new_df])
    index += 1    
```

### Grid Search

```python
def grid_search(parameters,regressor,algorithmName):
    
    from sklearn.model_selection import cross_val_score,GridSearchCV
    
    # add the parameters to the grid_search
    grid_search = GridSearchCV(
    estimator=regressor,
    param_grid=parameters,
    cv=5,
    n_jobs = -1 #opsiyonel
    )
    
    # fit the gird_search
    grid_search.fit(x_train,y_train)

    # calculate best parameters
    best_parameters = grid_search.best_params_

    # print parameters
    print(algorithmName.center(50,'*'))
    print(f'Best Parameters: {best_parameters}\n\n')    
```

### Random Forest

```python
def random_forest():
    from sklearn.ensemble import RandomForestRegressor
    regressor = RandomForestRegressor(n_estimators=100,random_state=0).fit(x_train,y_train)
    
    parameters = [{'n_estimators':[10,50,100]}]
    grid_search(parameters,regressor,'Random Forest')
    
    return regressor
```

### Decision Tree

```python
def decision_tree():
    from sklearn.tree import DecisionTreeRegressor
    regressor = DecisionTreeRegressor(random_state=100).fit(x_train,y_train)

    return regressor
```

### KNN

```python
def knn():
    from sklearn.neighbors import KNeighborsRegressor
    regressor = KNeighborsRegressor(n_neighbors=7).fit(x_train,y_train)
    
    parameters = [{'n_neighbors':[3,5,7]}]
    grid_search(parameters,regressor,'K Nearest Neighbors')
    
    return regressor
```

### Support Vector Machines

```python
def svm():
    from sklearn.svm import SVR
    regressor = SVR(kernel='poly',C = 5).fit(x_train,y_train)
    
    return regressor
```

### Multiple Linear Regression

```python
def multiple_lin_reg():
    from sklearn.linear_model import LinearRegression
    regressor = LinearRegression().fit(x_train,y_train)
    return regressor
```

### Polynomial Regression

```python
x_train_poly = 0
x_test_poly = 0

def polynomial_reg():
    from sklearn.preprocessing import PolynomialFeatures
    poly_reg = PolynomialFeatures(degree = 3)
    
    global x_train_poly,x_test_poly
    x_train_poly = poly_reg.fit_transform(x_train)
    x_test_poly = poly_reg.transform(x_test)
    
    from sklearn.linear_model import LinearRegression
    regressor = LinearRegression().fit(x_train_poly,y_train)
    
    return regressor
```

### XGBOOST

```python
def xgboost():
    from xgboost import XGBRegressor
    regressor = XGBRegressor().fit(x_train,y_train)
    return regressor
```

## Result

```python
df_result = pd.DataFrame()
index = 1
regressors = {'random_forest':random_forest(),'decision_tree':decision_tree(),'knn':knn(),'SVM':svm(),
             'multiple_lin_reg':multiple_lin_reg(),'poly_reg':polynomial_reg(),'xgboost':xgboost()}

for r in regressors:
    calculate_scores(regressors[r],r)
```
