## Data Preprocessing

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder # female = 0, male = 1
le = LabelEncoder()
X[:,1] = le.fit_transform(X[:,1])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X = np.array(X).astype("float32")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Creating ANN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import plot_model
```

```python
model = Sequential()

model.add(Dense(units=15,activation='relu'))

model.add(Dense(units=15,activation='relu'))

model.add(Dense(units=1,activation=None))

model.compile(optimizer='adam',loss='mse')
```

```python
early_stopping = EarlyStopping(mode = 'min',verbose=1,patience=50)
```

```python
model.fit(x=X_train,y = y_train,batch_size=32,epochs=300,validation_data=(X_test,y_test),callbacks=[early_stopping])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

## Plotting the Model

```python
plot_model(model,to_file = 'model.png')
```

## Result

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
y_pred = model.predict(X_test)

r2 = round(r2_score(y_test,y_pred),2)
mae = round(mean_absolute_error(y_test,y_pred),2)
mse = round(mean_squared_error(y_test,y_pred),2)

print(f'\nr2: {r2}\nmae: {mae}\nmse: {mse}')
```
